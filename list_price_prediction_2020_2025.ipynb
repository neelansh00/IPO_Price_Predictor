```python
combined_data['days_since_start'] = (combined_data['Date'] - combined_data['Date'].min()).dt.days
combined_data['Month']=combined_data['Date'].dt.month
combined_data['Day']=combined_data['Date'].dt.day
combined_data['DayOfWeek'] = combined_data['Date'].dt.dayofweek   #Monday=0, Sunday=6
combined_data['month_sin'] = np.sin(2 * np.pi * combined_data['Month'] / 12)
combined_data['month_cos'] = np.cos(2 * np.pi * combined_data['Month'] / 12)

# Day of month cyclical feature
# We assume max 31 days in a month
combined_data['day_sin'] = np.sin(2 * np.pi * combined_data['Day'] / 31)
combined_data['day_cos'] = np.cos(2 * np.pi * combined_data['Day'] / 31)

# Day of week cyclical feature
# We have 7 days in a week
combined_data['day_of_week_sin'] = np.sin(2 * np.pi * combined_data['DayOfWeek'] / 7)
combined_data['day_of_week_cos'] = np.cos(2 * np.pi * combined_data['DayOfWeek'] / 7)
combined_data['quarter'] = combined_data['Date'].dt.quarter
combined_data = pd.get_dummies(combined_data, columns=['quarter'], prefix='quarter',drop_first=True,dtype='int')

combined_data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import root_mean_squared_error,mean_absolute_error,r2_score,mean_absolute_percentage_error

cols=['IPO_size', 'Estimated_Price', 'QIB', 'NII', 'RII', 'Total',
       'Prev_Nifty_Open', 'Prev_Nifty_High', 'Prev_Nifty_Low',
       'Prev_Nifty_Close', 'Prev_Nifty_volume', 'Prev_Vix_Close',
       'Prev_Vix_change(%)', 'days_since_start', 'month_sin', 'month_cos',
       'day_sin', 'day_cos', 'day_of_week_sin', 'day_of_week_cos']
target='Listing_Price'


X=final_df[cols]
y=final_df[target]


# --- 3. Initialize TimeSeriesSplit for Cross-Validation ---
# n_splits: The number of splitting iterations in the cross-validator.
# A common choice is 5 or 10.
tscv = TimeSeriesSplit(n_splits=5) 

# --- 4. Prepare lists to store scores for each fold ---
dt_r2_scores = []
dt_mae_scores = []
dt_rmse_scores = []
dt_mape_scores = []

rf_r2_scores = []
rf_mae_scores = []
rf_rmse_scores = []
rf_mape_scores = []

xgb_r2_scores = []
xgb_mae_scores = []
xgb_rmse_scores = []
xgb_mape_scores = []

print("\nStarting TimeSeries Cross-Validation for each model...")

# --- 5. Loop through each fold and train/evaluate models ---
for fold, (train_index, test_index) in enumerate(tscv.split(X)):
    print(f"\n--- Fold {fold+1} ---")
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    print(f"  Training data size: {len(X_train)} samples")
    print(f"  Testing data size: {len(X_test)} samples")

    # --- Decision Tree Regressor ---
    dt_model = DecisionTreeRegressor(random_state=42)
    dt_model.fit(X_train, y_train)
    dt_y_pred = dt_model.predict(X_test)

    dt_r2 = r2_score(y_test, dt_y_pred)
    dt_mae = mean_absolute_error(y_test, dt_y_pred)
    dt_rmse = root_mean_squared_error(y_test, dt_y_pred)
    dt_mape = mean_absolute_percentage_error(y_test, dt_y_pred)

    dt_r2_scores.append(dt_r2)
    dt_mae_scores.append(dt_mae)
    dt_rmse_scores.append(dt_rmse)
    dt_mape_scores.append(dt_mape)
    print(f"  Decision Tree - R2: {dt_r2:.4f}, MAE: {dt_mae:.2f}, RMSE: {dt_rmse:.2f}, MAPE: {dt_mape:.4f}")

    # --- Random Forest Regressor ---
    rf_model = RandomForestRegressor(random_state=42)
    rf_model.fit(X_train, y_train)
    rf_y_pred = rf_model.predict(X_test)

    rf_r2 = r2_score(y_test, rf_y_pred)
    rf_mae = mean_absolute_error(y_test, rf_y_pred)
    rf_rmse = root_mean_squared_error(y_test, rf_y_pred)
    rf_mape = mean_absolute_percentage_error(y_test, rf_y_pred)

    rf_r2_scores.append(rf_r2)
    rf_mae_scores.append(rf_mae)
    rf_rmse_scores.append(rf_rmse)
    rf_mape_scores.append(rf_mape)
    print(f"  Random Forest - R2: {rf_r2:.4f}, MAE: {rf_mae:.2f}, RMSE: {rf_rmse:.2f}, MAPE: {rf_mape:.4f}")

    # --- XGBoost Regressor ---
    xgb_model = XGBRegressor(random_state=42, eval_metric='rmse') # eval_metric suppresses a warning
    xgb_model.fit(X_train, y_train)
    xgb_y_pred = xgb_model.predict(X_test)

    xgb_r2 = r2_score(y_test, xgb_y_pred)
    xgb_mae = mean_absolute_error(y_test, xgb_y_pred)
    xgb_rmse = root_mean_squared_error(y_test, xgb_y_pred)
    xgb_mape = mean_absolute_percentage_error(y_test, xgb_y_pred)

    xgb_r2_scores.append(xgb_r2)
    xgb_mae_scores.append(xgb_mae)
    xgb_rmse_scores.append(xgb_rmse)
    xgb_mape_scores.append(xgb_mape)
    print(f"  XGBoost - R2: {xgb_r2:.4f}, MAE: {xgb_mae:.2f}, RMSE: {xgb_rmse:.2f}, MAPE: {xgb_mape:.4f}")

# --- 6. Print Average Cross-Validation Scores ---
print("\n=== Average Cross-Validation Scores ===")

print("\nDecision Tree Averages:")
print(f"  Average R2: {np.mean(dt_r2_scores):.4f}")
print(f"  Average MAE: {np.mean(dt_mae_scores):.2f}")
print(f"  Average RMSE: {np.mean(dt_rmse_scores):.2f}")
print(f"  Average MAPE: {np.mean(dt_mape_scores):.4f}")

print("\nRandom Forest Averages:")
print(f"  Average R2: {np.mean(rf_r2_scores):.4f}")
print(f"  Average MAE: {np.mean(rf_mae_scores):.2f}")
print(f"  Average RMSE: {np.mean(rf_rmse_scores):.2f}")
print(f"  Average MAPE: {np.mean(rf_mape_scores):.4f}")

print("\nXGBoost Averages:")
print(f"  Average R2: {np.mean(xgb_r2_scores):.4f}")
print(f"  Average MAE: {np.mean(xgb_mae_scores):.2f}")
print(f"  Average RMSE: {np.mean(xgb_rmse_scores):.2f}")
print(f"  Average MAPE: {np.mean(xgb_mape_scores):.4f}")
```
